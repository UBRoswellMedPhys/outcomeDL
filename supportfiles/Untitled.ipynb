```python
import numpy as np
import os
```

```python
folder = r"D:\extracteddata"
patients = [x for x in os.listdir(folder) if os.path.isdir(os.path.join(folder,x))]
```

```python
for patient in patients:
    subpath = os.path.join(folder,patient)
    mask_l = np.load(os.path.join(subpath,'parotid_l_mask.npy'))
    mask_r = np.load(os.path.join(subpath,'parotid_r_mask.npy'))
    mask = mask_l + mask_r
    if np.sum(mask) == 0:
        continue
    ind = np.argwhere(mask)
    z = np.amax(ind[:,0]) - np.amin(ind[:,0])
    y = np.amax(ind[:,1]) - np.amin(ind[:,1])
    x = np.amax(ind[:,2]) - np.amin(ind[:,2])
    print(patient, z, y, x)
```

```python
com = np.sum(ind,axis=0) / len(ind)
```
